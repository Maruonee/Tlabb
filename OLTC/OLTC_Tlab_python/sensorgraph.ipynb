```python
import re
import matplotlib.pyplot as plt

def extract_vr1_values(file_path):
    vr1_values = []
    with open(file_path, 'r') as file:
        for line in file:
            match = re.search(r'VR1 : (\d+)', line)
            if match:
                vr1_value = int(match.group(1))
                vr1_values.append(vr1_value)
    return vr1_values

def plot_vibration(vr1_values):
    plt.figure(figsize=(10, 5))
    plt.plot(vr1_values, marker='o', linestyle='-', markersize=3, linewidth=0.8)
    plt.title("Vibration of VR1 Sensor")
    plt.xlabel("Time (seconds)")
    plt.ylabel("VR1 Value")
    plt.grid(True)
    plt.show()

# 파일 경로 설정
file_path = 'C:\\Users\\tlab\\Desktop\\241031_2_sensors\\241031_2_sensors_20241031_153949.txt'

# VR1 값 추출
vr1_values = extract_vr1_values(file_path)

# 데이터 시각화
plot_vibration(vr1_values)
```

```python
# 필요한 라이브러리 임포트
import re
import matplotlib.pyplot as plt

# 파일 경로 설정
file_path1 = 'C:\\Users\\tlab\\Desktop\\241106_1_sensors_20241106_170453.txt'
file_path2 = 'C:\\Users\\tlab\\Desktop\\241017_1_sensors_20241106_170419.txt'

# VR1 값을 추출하는 함수 정의
def extract_vr1_values(file_path):
    vr_values = []
    with open(file_path, 'r') as file:
        for line in file:
            # 정규 표현식을 사용하여 'VR1 :' 뒤의 숫자 추출
            match = re.search(r'VR1 : (\d+)', line)
            if match:
                vr_value = int(match.group(1))  # 숫자 값을 정수로 변환
                vr_values.append(vr_value)
    return vr_values

# 각 파일에서 VR1 값 추출
vr1_values = extract_vr1_values(file_path1)
vr2_values = extract_vr1_values(file_path2)

# 그래프 생성
plt.figure(figsize=(12, 6))
plt.plot(vr1_values, label="Yulchon", linewidth=0.1)
plt.plot(vr2_values, label="Buheung", linewidth=0.1)
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# 이미지 로드
image_path = 'C:\\Users\\tlab\\Desktop\\pic.jpg'  # 업로드한 이미지 경로
image = Image.open(image_path)
image = np.array(image)

# R, G, B 채널 분리하여 각각 색상 강조
R_channel = np.zeros_like(image)
G_channel = np.zeros_like(image)
B_channel = np.zeros_like(image)

R_channel[:, :, 0] = image[:, :, 0]  # R 채널만 남기고 나머지는 0
G_channel[:, :, 1] = image[:, :, 1]  # G 채널만 남기고 나머지는 0
B_channel[:, :, 2] = image[:, :, 2]  # B 채널만 남기고 나머지는 0

# 시각화
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(R_channel)
plt.title('Red Channel')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(G_channel)
plt.title('Green Channel')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(B_channel)
plt.title('Blue Channel')
plt.axis('off')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from PIL import Image

# 이미지 로드
image_path = 'C:\\Users\\tlab\\Desktop\\pic.jpg'  # 업로드한 이미지 경로
image = Image.open(image_path)
image = np.array(image)

# R, G, B 채널 추출
R_channel = image[:, :, 0]
G_channel = image[:, :, 1]
B_channel = image[:, :, 2]

# x, y 좌표 생성
x = np.arange(R_channel.shape[1])
y = np.arange(R_channel.shape[0])
x, y = np.meshgrid(x, y)

# 3D 시각화
fig = plt.figure(figsize=(15, 10))

# R 채널
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(x, y, R_channel, cmap='Reds', edgecolor='none')
ax1.set_title('Red Channel Intensity')
ax1.set_xlabel('Width (pixels)')
ax1.set_ylabel('Height (pixels)')
ax1.set_zlabel('Intensity')

# G 채널
ax2 = fig.add_subplot(132, projection='3d')
ax2.plot_surface(x, y, G_channel, cmap='Greens', edgecolor='none')
ax2.set_title('Green Channel Intensity')
ax2.set_xlabel('Width (pixels)')
ax2.set_ylabel('Height (pixels)')
ax2.set_zlabel('Intensity')

# B 채널
ax3 = fig.add_subplot(133, projection='3d')
ax3.plot_surface(x, y, B_channel, cmap='Blues', edgecolor='none')
ax3.set_title('Blue Channel Intensity')
ax3.set_xlabel('Width (pixels)')
ax3.set_ylabel('Height (pixels)')
ax3.set_zlabel('Intensity')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from PIL import Image

# 이미지 로드
image_path = 'C:\\Users\\tlab\\Desktop\\pic.jpg'  # 업로드한 이미지 경로
image = Image.open(image_path)
image = np.array(image)

# R, G, B 채널 추출
R_channel = image[:, :, 0]
G_channel = image[:, :, 1]
B_channel = image[:, :, 2]

# x, y 좌표 생성
x = np.arange(R_channel.shape[1])
y = np.arange(R_channel.shape[0])
x, y = np.meshgrid(x, y)

# 3D 선 그래프 시각화
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d', facecolor='white')

# R, G, B 채널 각각에 대해 선 그래프를 그림
# 각 채널의 x, y 위치에 해당하는 픽셀 강도를 연결하여 선을 그립니다.
for i in range(0, R_channel.shape[0], 10):  # 10 픽셀 간격으로 샘플링하여 표시
    ax.plot(x[i, :], y[i, :], R_channel[i, :], color='red', linewidth=0.5, alpha=0.7)
    ax.plot(x[i, :], y[i, :], G_channel[i, :], color='green', linewidth=0.5, alpha=0.7)
    ax.plot(x[i, :], y[i, :], B_channel[i, :], color='blue', linewidth=0.5, alpha=0.7)

# 그래프 회전 각도 설정
ax.view_init(elev=30, azim=120)

# 축 라벨 및 제목 설정
ax.set_xlabel('Width (pixels)', fontsize=12, labelpad=10)
ax.set_ylabel('Height (pixels)', fontsize=12, labelpad=10)
ax.set_zlabel('Intensity', fontsize=12, labelpad=10)
ax.set_title('3D Line Plot of RGB Channel Intensity', fontsize=14, pad=20)

# 범례
ax.plot([], [], color='red', label='Red Channel', linewidth=1.5)
ax.plot([], [], color='green', label='Green Channel', linewidth=1.5)
ax.plot([], [], color='blue', label='Blue Channel', linewidth=1.5)
ax.legend(loc='upper right', fontsize=10)

plt.show()
```
