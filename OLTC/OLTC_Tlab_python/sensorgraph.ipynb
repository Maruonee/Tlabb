```python
# 셀 1: 필요한 라이브러리 임포트
import re
import matplotlib.pyplot as plt

# 셀 2: 데이터 파일 경로 설정 및 VR1 값 추출
file_path = 'C:\\Users\\tlab\\Desktop\\241031_2_sensors\\241031_2_sensors_20241031_153949.txt'
vr1_values = []

with open(file_path, 'r') as file:
    for line in file:
        # 정규 표현식을 사용하여 'VR1 :' 뒤의 숫자 추출
        match = re.search(r'VR1 : (\d+)', line)
        if match:
            vr1_value = int(match.group(1))  # 숫자 값을 정수로 변환
            vr1_values.append(vr1_value)

# 추출한 값 확인
print(len(vr1_values))
print(f"second {len(vr1_values)/91}")
```

```python
# 셀 3: VR1 값 그래프 시각화
plt.figure(figsize=(10, 5))
plt.plot(vr1_values, marker='o', linestyle='-', markersize=3, linewidth=0.8)
plt.title("Vibration")
plt.xlabel("time")
plt.ylabel("VR1 Value")
plt.grid(True)
plt.show()
```

```python
import re
import matplotlib.pyplot as plt

def extract_vr1_values(file_path):
    vr1_values = []
    with open(file_path, 'r') as file:
        for line in file:
            match = re.search(r'VR1 : (\d+)', line)
            if match:
                vr1_value = int(match.group(1))
                vr1_values.append(vr1_value)
    return vr1_values

def plot_vibration(vr1_values):
    plt.figure(figsize=(10, 5))
    plt.plot(vr1_values, marker='o', linestyle='-', markersize=3, linewidth=0.8)
    plt.title("Vibration of VR1 Sensor")
    plt.xlabel("Time (seconds)")
    plt.ylabel("VR1 Value")
    plt.grid(True)
    plt.show()

# 파일 경로 설정
file_path = 'C:\\Users\\tlab\\Desktop\\241031_2_sensors\\241031_2_sensors_20241031_153949.txt'

# VR1 값 추출
vr1_values = extract_vr1_values(file_path)

# 데이터 시각화
plot_vibration(vr1_values)
```
